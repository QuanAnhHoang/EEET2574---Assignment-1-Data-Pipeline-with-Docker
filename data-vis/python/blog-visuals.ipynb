```python
weather_data = getWeatherDF()
weather_data
```

```python
# Convert 'forecastdate' column to datetime format
weather_data['forecastdate'] = pd.to_datetime(weather_data['forecastdate'])

# Filter data for Ho Chi Minh City
hcm_data = weather_data[weather_data['location'] == 'Ho Chi Minh City']
avg_hcm_temp= (hcm_data['temp_max']+hcm_data['temp_min'])/2
# Filter data for Tokyo
tokyo_data = weather_data[weather_data['location'] == 'Tokyo']
avg_tokyo_temp= (tokyo_data['temp_max']+tokyo_data['temp_min'])/2

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(hcm_data['forecastdate'], avg_hcm_temp, label='HCM AVG Temp', color='blue', marker='o')
plt.plot(hcm_data['forecastdate'], hcm_data['feels_like'], label='HCM feels_like Temp', color='lightcoral', marker='o')

plt.plot(tokyo_data['forecastdate'], avg_tokyo_temp, label='Tokyo AVG Temp', color='red', marker='o')
plt.plot(tokyo_data['forecastdate'], tokyo_data['feels_like'], label='Tokyo feels_like Temp', color='green', marker='o')

plt.title('Average Temperature and Feels-like temperature comparison - Ho Chi Minh City vs. Tokyo')
plt.xlabel('Date and Time')
plt.ylabel('Temperature (K)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
fl_data = getFlightLogsDF()
fl_data
```

```python
fl_data['route'] = fl_data['departure_airport'] + ' to ' + fl_data['arrival_airport']

# Find the top 10 busiest routes according to 'passenger_count'
top_10_routes_passenger_count = fl_data.groupby('route')['passenger_count'].sum().nlargest(10)

# Plotting
plt.figure(figsize=(10, 6))
top_10_routes_passenger_count.plot(kind='bar', color='skyblue')
bars = top_10_routes_passenger_count.plot(kind='bar', color='skyblue')

for bar in bars.patches:
    plt.annotate(f'{bar.get_height():,.0f}', 
                 (bar.get_x() + bar.get_width() / 2, bar.get_height()), 
                 ha='center', va='bottom', fontsize=8, color='black')
    
plt.title('Top 10 Busiest Routes by Passenger Count')
plt.xlabel('Route')
plt.ylabel('Total Passenger Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
top_3_airlines = fl_data['airline'].value_counts().nlargest(3)

# Plotting
plt.figure(figsize=(8, 5))
bars = top_3_airlines.plot(kind='bar', color='lightcoral')

for bar in bars.patches:
    plt.annotate(f'{bar.get_height():,.0f}', 
                 (bar.get_x() + bar.get_width() / 2, bar.get_height()), 
                 ha='center', va='bottom', fontsize=8, color='black')
    
plt.title('Top 3 Most Common Airlines')
plt.xlabel('Airline')
plt.ylabel('Number of Flights')
plt.xticks(rotation=45, ha='right')
plt.show()
```
